### Few-Shot vs. Zero-Shot Prompting
#### scenario: Predict whether an email is urgent based on its content.

### Zero-Shot Prompt
#### In zero-shot learning, the model is given the task without any prior examples:

Prompt:<br>
```"Classify this sentence: 'The new smartphone model features a 120Hz display, improved camera, and longer battery life.'"```
```Expected Outcome: Technology```
``` Model Response: Smartphones ```



### Few-Shot Prompt
#### In few-shot learning, the model is given a few examples first, which helps it understand the context and nuances better: 

Prompt:<br>
```
Classify each sentence into one of the categories: "Sports," "Technology," or "Entertainment."

1. "The football team won the championship after an exciting match." – Sports
2. "The latest Marvel movie has broken box office records this weekend." – Entertainment
3. "The new gaming laptop comes with an RTX 3080 graphics card and 4K display." – Technology

Now classify this sentence: "The new smartphone model features a 120Hz display, improved camera, and longer battery life."
```
<br>
 Model Response: ```Technology```
<hr>

Effectiveness: The few-shot example helps the model grasp the context by showing it examples of both urgent and non-urgent communications, 
    thereby improving accuracy in classifying new emails.

### Chain-of-Thought Prompting
```Calculate the total price after a discount and tax on an item.```<br>
Without Chain-of-Thought
Prompt:<br>
```A jacket costs $120, and there's a 25% discount available. After the discount, a 7% sales tax is applied. What is the final price?
Model Response: Direct calculation without explanation.
```


With Chain-of-Thought
Prompt:<br>
```
Let’s calculate it step-by-step: First, find the price after the 25% discount. Discount amount = 25% of $120 = $30. So,
 discounted price = $120 - $30 = $90. Now, apply a 7% sales tax to the discounted price. Tax = 7% of $90 = $6.30. 
 Finally, add the tax to the discounted price to get the final amount: Final price = $90 + $6.30.
```

```Model Response:
First, we calculate the discount on the jacket: 25% of $120 = $30. So, the price after the discount is $120 - $30 = $90.
 Next, we need to add sales tax to the discounted price. The sales tax is 7% of $90, which is $6.30. Adding this to the discounted price, the final price of the jacket becomes $90 + $6.30 = $96.30.
```
<hr>
Effectiveness: The chain-of-thought prompting clearly outlines the intermediate steps, leading to a more understandable and traceable calculation process.<br> 
    This method promotes transparency in the model’s reasoning, making it easier to follow and verify.

### Conclusion
Zero-shot and few-shot prompting are powerful techniques in prompting models to perform tasks with or without prior examples.<br>
    Chain-of-thought prompting enhances the transparency and traceability of the model's reasoning by breaking down complex tasks into step-by-step calculations or explanations.<br>
    These prompting strategies can significantly improve the performance and interpretability of AI models in various applications.

````python
## 12 

def fibonacci(n):
    fib = [0, 1] + [0]*(n-1)
    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]
    return fib[n]
    
import random
n = random.randint(1, 20000)
result = fibonacci(n)
print(f"The {n}-th term in the Fibonacci series is: {result}")
````

````python
## 13 

def find_longest_common_substring(str1, str2):
    n, m = len(str1), len(str2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    max_length = 0
    end_index_str1 = 0

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
                if dp[i][j] > max_length:
                    max_length = dp[i][j]
                    end_index_str1 = i
            else:
                dp[i][j] = 0
    longest_substring = str1[end_index_str1 - max_length: end_index_str1]
    return longest_substring


# string1 = "We are currently in lockdown"
# string2 = "There was a lock on the door"
string1= "There is a lady on the mountain with an umbrella"
string2= "I love to look at the mountain when it is snowcapped"
longest_substring = find_longest_common_substring(string1, string2)
longest_substring
````

````python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2, mutual_info_classif
from sklearn.preprocessing import LabelEncoder
from scipy.stats import chi2_contingency


categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
encoders = {}
for column in categorical_columns:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
    encoders[column] = encoder

if 'Decision' in df.columns and df['Decision'].dtype == 'object':
    encoder = LabelEncoder()
    df['Decision'] = encoder.fit_transform(df['Decision'])
    encoders['Decision'] = encoder

chi_scores, p_vals = chi2(df[categorical_columns], df['Decision'])
chi_results = pd.DataFrame({'Feature': categorical_columns, 'Chi-Square Score': chi_scores, 'P-value': p_vals})

print("Chi-Square Test of Independence Results:")
print(chi_results)

mi_scores = mutual_info_classif(df[categorical_columns], df['Decision'], discrete_features=True)
mi_results = pd.DataFrame({'Feature': categorical_columns, 'Mutual Information Score': mi_scores})

print("\nMutual Information Results:")
print(mi_results)
````

````python
# WORKING WITH MISSING DATA
# IGNORE WARNING IMPORT 
import warnings
warnings.filterwarnings('ignore')

employment_info_1_mean= data['Insurance_History_5'].mean()
data_filled = data.copy()
data_filled['Insurance_History_5'].fillna(employment_info_1_mean, inplace=True)

print(f"Median value used {employment_info_1_mean}")
print(f"Employment_Info_1 after filling: {data_filled['Insurance_History_5'].isnull().sum()}")
````

````python
# DROP ALL OTHER NULL VALS
data_no_nulls = data_filled.dropna(axis=1)

# Checking the result
print(f"Columns dropped: {set(data.columns) - set(data_no_nulls.columns)}")
print(f"New DataFrame shape: {data_no_nulls.shape}")
````

````python
# Exploring Multicollinearity in the new dataframe with handled missing values 
# !pip install seaborn
# !pip install statsmodels

import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

data_no_nulls = pd.get_dummies(data_no_nulls, columns=['Product_Info_2'])

# Convert all values to float
for col in data_no_nulls.columns:
    if data_no_nulls[col].dtype == 'object':
        data_no_nulls[col] = data_no_nulls[col].astype('category').cat.codes
    data_no_nulls[col] = data_no_nulls[col].astype(float)

correlation_matrix = data_no_nulls.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

X = data_no_nulls.drop(columns=['Response'])
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("Variance Inflation Factor (VIF):")
print(vif_data)
````

````python
from sklearn.feature_selection import mutual_info_regression
from sklearn.ensemble import RandomForestRegressor

# Define predictors and target variable
predictors = data_no_nulls.drop(columns=['Id', 'Response'])
target = data_no_nulls['Response']

# Calculate Mutual Information scores
mi_scores = mutual_info_regression(predictors, target)
mi_sorted = pd.Series(mi_scores, index=predictors.columns).sort_values(ascending=False)
top_features_by_mi = mi_sorted.head(3).index.tolist()
print("Top 3 predictors determined by Mutual Information:", top_features_by_mi)

# Evaluate feature importance using Random Forest
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(predictors, target)
importances_rf = pd.Series(rf_regressor.feature_importances_, index=predictors.columns).sort_values(ascending=False)
top_features_by_rf = importances_rf.head(3).index.tolist()
print("Top 3 predictors determined by Random Forest Importance:", top_features_by_rf)
````
